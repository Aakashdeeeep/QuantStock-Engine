```python
import pandas as pd 
import numpy as np 
import yfinance as yf 
import math
```

Load Stock List

```python
tickers = pd.read_csv('top_50_indian_stocks.csv')
tickers.head()
```

```python
def fetch_market_cap(tickers_list):
    data = yf.download(tickers_list,period='1d',group_by='ticker',auto_adjust='False')
    stocks_data = []
    for ticker in tickers_list:
        lastest_price = data[ticker]['Close'].iloc[-1]
        market_cap = yf.Ticker(ticker).info.get("marketCap",'N/A')
        stocks_data.append({
            "Ticker" : ticker,
            "Market Cap" : market_cap,
            "Latest Price" : lastest_price
        })
    stocks_df = pd.DataFrame(stocks_data)
    return stocks_df 
        
tickers_list = tickers["Ticker"].values.tolist()
df = fetch_market_cap(tickers_list)
df.head()
    
    
```

```python
df
```

```python
df = df.sort_values(by = "Market Cap",ascending=False)
df
```
